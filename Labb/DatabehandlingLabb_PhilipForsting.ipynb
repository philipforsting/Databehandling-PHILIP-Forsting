```python
#f
df_college["Elite"] = pd.cut(df_college["Top10perc"], [0, 50, 100], labels = ["No", "Yes"])
df_college.value_counts("Elite")
df_college.boxplot(column="Outstate", by="Elite")
```

```python
#g
fig, axs = plt.subplots(2, 2, figsize=(12, 8))             # inspiration taken from https://matplotlib.org/stable/gallery/subplots_axes_and_figures/subplots_demo.html

df_college["Top10perc"].plot.hist(ax=axs[0, 0])
axs[0, 0].set_title("Top10perc")
df_college["Top25perc"].plot.hist(ax=axs[0, 1])
axs[0, 1].set_title("Top25perc")
df_college["PhD"].plot.hist(ax=axs[1, 0])
axs[1, 0].set_title("PhD")
df_college["Grad.Rate"].plot.hist(ax=axs[1, 1])
axs[1, 1].set_title("Grad.Rate")
```

### What I am looking for
I have chosen to look at the parameters suggested in the exercise: "are there any correlations between personal spending, instructional spending and Elite status". I took the oppurtiunity to experiment with different plotting methods: 
* The Seaborne Scatter plot visualise the relations between at all three categories with Elite as hue.  
* The Scatter Matrix visualise the relations between at all three categories (Category Top10perc was chosen insted of Elite status since Elite does not contain numerical values).
* The box plots look at the relation between personal spending and instructional spending in relation to Elite status.

### What I find
It apears to be a correlation between hight amount of Instructional expenditure per student and Elite status, clearly visualised all plots. Personaly, I find the boxplot to have the clearest visualisation.
