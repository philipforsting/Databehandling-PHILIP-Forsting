```python
#  e) Find all rows where the freetime is NaN.
dfstudents_freetime_NaN_mask = df_students["freetime"].isna() # Na-mask for column "freetime". True if value is Na
dfstudents_freetime_NaN = df_students[dfstudents_freetime_NaN_mask] # pass dataframe through mask. Dataframe rows existing in mask will pass
dfstudents_freetime_NaN_indexes = dfstudents_freetime_NaN.index # returns series containing indexes as reruested
dfstudents_freetime_NaN_indexes
```

```python
# f) Find all rows where the freetime or the age is NaN.
students_freetimeOrAge_NaN_mask = df_students["freetime"].isna() | df_students["age"].isna() # Na-mask for column "freetime". True if value is Na
df_students_freetimeOrAge_NaN = df_students[students_freetimeOrAge_NaN_mask] # pass dataframe through mask. Dataframe rows existing in mask will pass
students_freetimeOrAge_NaN_indexes = df_students_freetimeOrAge_NaN.index
students_freetimeOrAge_NaN_indexes
```

```python
#  g) You will notice that some rows have several NaNs. 
# Now compute the proportion that these rows constitute of the whole dataset.
df_students_NaN = df_students.isna() # returns dataframe, true/false if cell contains Nan
s_students_NaN_sum = df_students_NaN.sum(axis=1)  # Returns series, value represents number of Nan per row
s_students_NaN_sum_gt1 = s_students_NaN_sum > 1 # Mask contain true if number of Nan per row is gt 1

df_students_NaN_sum_gt1_sum = s_students_NaN_sum_gt1.sum() # sums series
df_students_NaN_sum_gt1_sum
```

```python
#  a) As you have conversed with a domain expert you both agree that there are too many missing 
# data to fill in and the proportion is small enough to be safe to just remove. 
# Now remove these rows and use your missing-value utility function visualize the remaining NaNs.

df_students_drop_multNaN_index = df_students[s_students_NaN_sum_gt1].index # Mask contain true if number of Nan per row is gt 1
df_students_drop_multNaN = df_students.drop(df_students_drop_multNaN_index)
data_utils.NanPlotter(df_students_drop_multNaN)
```
