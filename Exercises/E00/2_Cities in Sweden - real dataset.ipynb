```python
# visualization using seaborn
fig, axes = plt.subplots(1,2, dpi=120, figsize=(16,4))

df_cities = df_cities.sort_values(by="Folkmängd 2020", ascending=False)
df_cities_top5 = df_cities.sort_values(by="Folkmängd 2020", ascending=False).iloc[:5]
df_cities_low5 = df_cities.sort_values(by="Folkmängd 2020", ascending=False).iloc[-5:]

titles = ["Sveriges 5 största kommuner 2020", "Sveriges 5 minsta kommuner 2020"]
data_frames = [df_cities_top5, df_cities_low5]
x_column = ["Kommun", "Kommun"]

for i, (data, title) in enumerate(zip(data_frames, titles)): 
    sns.barplot(data = data, x=x_column[i], y="Folkmängd 2020", ax = axes[i])
    axes[i].set(title=title)
    axes[i].set_xticklabels(axes[i].get_xticklabels(), rotation=90)
plt.show()
```
